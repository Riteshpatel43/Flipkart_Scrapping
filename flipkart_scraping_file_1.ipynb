```python
 from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
from bs4 import BeautifulSoup as bs

 
driver = webdriver.Chrome()
url = "https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles&requestId=c4adac3a-c2a2-4650-a78b-3008f6b1d377&page=1"
driver.get(url)

# list to store data
mobile_data = []

# Function to scrape data from the current page
def scrape_page():
    soup = bs(driver.page_source, 'html.parser')   
    mobiles = soup.find_all('div', attrs={'class': 'cPHDOP col-12-12'})
    details = soup.find_all('div', attrs={'class': '_6NESgJ'})
    
    for mobile in mobiles:
        
        name_element = mobile.find('div', attrs={'class': 'KzDlHZ'})
        price_element = mobile.find('div', attrs={'class': 'Nx9bqj _4b5DiR'})
        rating_element = mobile.find('div',attrs={'class':'XQDdHH'})
        rating_count_element = mobile.find('span',attrs={'class':'Wphh3N'})
        ram_element = mobile.find('div',attrs={'class':'_6NESgJ'})
        
        name = name_element.get_text(strip=True) if name_element else 'N/A'
        price = price_element.get_text(strip=True) if price_element else 'N/A'
        rating = rating_element.get_text(strip=True) if rating_element else 'N/A'
        rating_count = rating_count_element.get_text(strip=True) if rating_count_element else 'N/A'
         
        
        # for ram ========================================================================        
        if ram_element:
            # Find the <ul> inside the ram_element
            ul_element = ram_element.find('ul', attrs={'class': 'G4BRas'})
            
            # If ul exists, find the first li
            if ul_element:
                first_li = ul_element.find('li', attrs={'class': 'J+igdf'})
                
                # If the first li exists, get its text
                if first_li:
                    ram = first_li.get_text(strip=True)
                else:
                    ram = 'N/A'  # Handle case where li is not found
            else:
                ram = 'N/A'  # Handle case where ul is not found
        else:
            ram = 'N/A'  # Handle case where ram_element is not found
         # for ram ========================================================================   

         # for display ========================================================================   
                
        if ram_element:
            # Find the ul inside the ram_element
            ul_element = ram_element.find('ul', attrs={'class': 'G4BRas'})
            
            # If ul exists, find all li elements within it
            if ul_element:
                li_elements = ul_element.find_all('li', attrs={'class': 'J+igdf'})  # Find all li elements
                
                # If there are at least two li elements, get the second one (index 1)
                if len(li_elements) > 1:
                    second_li = li_elements[1]  # Select the second li
                    display_details = second_li.get_text(strip=True)  # Extract the text from the second li
                else:
                    display_details = 'N/A'  # Handle case where there is no second li
            else:
                display_details = 'N/A'  # Handle case where ul is not found
        else:
            display_details = 'N/A'  # Handle case where ram_element is not found
         # for display ========================================================================  
    
        # for camera ========================================================================   
                
        if ram_element:
            # Find the ul inside the ram_element
            ul_element = ram_element.find('ul', attrs={'class': 'G4BRas'})
            
            # If <ul> exists, find all li elements within it
            if ul_element:
                li_elements = ul_element.find_all('li', attrs={'class': 'J+igdf'})  # Find all li elements
                
                # If there are at least two li elements, get the second one (index 1) for camera
                if len(li_elements) > 2:
                    camera_li = li_elements[2]  # Select the second li (usually camera)
                    camera_details = camera_li.get_text(strip=True)  # Extract the text from the second li
                else:
                    camera_details = 'N/A'  # Handle case where there is no second li
            else:
                camera_details = 'N/A'  # Handle case where ul is not found
        else:
            camera_details = 'N/A'  # Handle case where ram_element is not found

         # for camera ========================================================================  

         # for battery ========================================================================   
              
        if ram_element:
            # Find the ul inside the ram_element
            ul_element = ram_element.find('ul', attrs={'class': 'G4BRas'})
            
            # If ul exists, find all <li> elements within it
            if ul_element:
                li_elements = ul_element.find_all('li', attrs={'class': 'J+igdf'})  # Find all li elements
                
                # If there are at least two li elements, get the second one (index 1) for camera
                if len(li_elements) > 3:
                    battery_li = li_elements[3]  # Select the second li (usually camera)
                    battery_details = battery_li.get_text(strip=True)  # Extract the text from the second li
                else:
                    battery_details = 'N/A'  # Handle case where there is no second li
            else:
                battery_details = 'N/A'  # Handle case where ul is not found
        else:
            battery_details = 'N/A'  # Handle case where ram_element is not found

         # for battery ========================================================================  
        
         # for battery ========================================================================   
               
        if ram_element:
            # Find the ul inside the ram_element
            ul_element = ram_element.find('ul', attrs={'class': 'G4BRas'})
            
            # If <ul> exists, find all li elements within it
            if ul_element:
                li_elements = ul_element.find_all('li', attrs={'class': 'J+igdf'})  # Find all li elements
                
                # If there are at least two li elements, get the second one (index 1) for camera
                if len(li_elements) > 4:
                    processor_li = li_elements[4]  # Select the second li (usually camera)
                    processor_details = processor_li.get_text(strip=True)  # Extract the text from the second li
                else:
                    processor_details = 'N/A'  # Handle case where there is no second li
            else:
                processor_details = 'N/A'  # Handle case where ul is not found
        else:
            processor_details = 'N/A'  # Handle case where ram_element is not found

         # for battery ========================================================================  

        
        # Appending data to list
        mobile_data.append({'Name': name, 'Price': price,'Rating':rating,'Rating_count':rating_count,
                            'ram':ram,'Display': display_details ,'cameras':camera_details,'battery':battery_details,'processor':processor_details})

# Pagination logic
page_number = 1
while True:
    print(f"Scraping Page {page_number}")
    scrape_page()  # Scrape the current page

    try:
        # if we have click page_number or next button
        if page_number % 10 == 0:
            next_button = driver.find_element(By.XPATH, "//span[text()='Next']")
            next_button.click()
            time.sleep(3)
        else:
            # clicking next page number
            next_page = driver.find_element(By.XPATH, f"//a[text()='{page_number + 1}']")
            next_page.click()
            time.sleep(15)
        
        page_number += 1  # incrementing page count
    except:
        print("No more pages to scrape.")
        break

 
df = pd.DataFrame(mobile_data)
# print(df)

# Close the driver
# driver.quit()
```
